```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt

def nothing(*arg):
    pass

filename = 'v3.avi'
cap = cv2.VideoCapture(filename)

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
center = int (width/2) - 50
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
length = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

j = 0
n = 1
winHeight = 128
winWidth = 308

y = int ((height- winHeight)/2) 
yy = y + winHeight - 1
x1 = center - int (3*winWidth/2)
x2 = center - int (winWidth/2)
x3 = center + int (winWidth/2)
x11 = x1 + winWidth -1
x22 = x2 + winWidth -1
x33 = x3 + winWidth -1

print("x1=",x1,"x11=",x11,"x2=",x2,"x22=",x22,"x3=",x3,"x33=",x33)

while (cap.isOpened()):

    ret, frame = cap.read()

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    img = cv2.GaussianBlur(gray, (5, 5), 1.5)

    crop1 = img[y:yy,x1:x11]
    crop2 = img[y:yy,x2:x22]
    crop3 = img[y:yy,x3:x33]
 
    s = 'tile1/tile('+str(n)+r').jpg'
    cv2.imwrite(s,crop1)

    s = 'tile2/tile('+str(n)+r').jpg'
    cv2.imwrite(s,crop2)

    s = 'tile3/tile('+str(n)+r').jpg'
    cv2.imwrite(s,crop3)

    n = n + 1
    j = j + 1

    #Вывод изображений
    if j == 10 :    
      cv2_imshow(crop3)
      j = 0

    key = cv2.waitKey(1) & 0xFF
    if key == ord('q'):
        break
    if n>100000 :    
        break

cap.release()
```
